```python
from vllm import LLM, SamplingParams
llm = LLM(
    "tcz/rb-llama-90b",                 # your checkpoint
    tensor_parallel_size=4,   # one shard per H100
    dtype="bfloat16",

    # --- keep the next four lines until the model loads ---
    gpu_memory_utilization=0.9,  # reserve 85 % instead of 90 %
    enforce_eager=True,           # skip CUDA-Graph → smaller peak
    max_model_len=12_000,           # 4 k context; raise later if you need
    max_num_seqs=64,              # concurrent sequences
)
```

```python
data_prompt = """Your job is to take an SVG file of a web design and convert it into a pixel-perfect HTML and CSS markup and stylesheet.

### Input:
{}

### Response:
{}"""
```

```python
max_tokens = 50_000
sampling_params = SamplingParams(
    temperature   = 0.0,
    top_p         = 1.0,
    top_k         = 1,
    n             = 1,
    max_tokens    = max_tokens,
    seed          = 0
)
```

```python
import os 

tokenizer = llm.get_tokenizer()
prompts = []

for filename in os.listdir('/svgs'):
    file_path = os.path.join('/svgs', filename)
    if os.path.isfile(file_path):
        with open(file_path, 'r') as file:
            content = file.read()
            prompt = data_prompt.format(
                content,
                "",
            )
        
            tokenized_ids = tokenizer(prompt).input_ids
            if len(tokenized_ids) > max_tokens:
                print(f"Chopped input from {len(tokenized_ids)}")
                prompt = tokenizer.decode(tokenized_ids[:max_tokens])
        
            prompts.append(prompt)

print(len(prompts))
print(prompts[0][:200])
```

```python
from tqdm import tqdm

htmls = []

for prompt in tqdm(prompts):
    output = llm.generate([prompt], sampling_params)
    htmls.append(output[0].outputs[0].text)
    # print(output)

import pickle

with open('generated_htmls.pkl', 'wb') as f:
    pickle.dump(htmls, f)
```

```python
high_temp_sampling_params = SamplingParams(
    temperature = 0.8,   
    top_p       = 0.95,  
    max_tokens  = max_tokens,
    n           = 5,
    seed        = 0,
)

from tqdm import tqdm

htmls = []

for prompt in tqdm(prompts):
    output = llm.generate([prompt], high_temp_sampling_params)

    html_variants = []
    for o in output[0].outputs:
        html_variants.append(o.text)

    htmls.append(html_variants)

import pickle

with open('generated_htmls_high_temp.pkl', 'wb') as f:
    pickle.dump(htmls, f)

print(htmls)
```
