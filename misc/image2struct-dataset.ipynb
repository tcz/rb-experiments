```python
from datasets import load_dataset, concatenate_datasets
dataset = concatenate_datasets([
    load_dataset('stanford-crfm/image2struct-webpage-v1', 'css')['validation'],
    load_dataset('stanford-crfm/image2struct-webpage-v1', 'javascript')['validation'],
    load_dataset('stanford-crfm/image2struct-webpage-v1', 'html')['validation'],
])

dataset
```

```python
import os
import tarfile
import base64

def extract_repo(base64_encoding: str, repo_path: str) -> None:
    # Decode the base64 string which corresponds to an archive
    # and extract the files to the repo_path
    try:
        archive = base64.b64decode(base64_encoding)
        # Write to .tar file
        with open(os.path.join(repo_path, "repo.tar.gz"), "wb") as f:
            f.write(archive)
        # Extract
        with tarfile.open(os.path.join(repo_path, "repo.tar.gz"), "r:gz") as tar:
            tar.extractall(repo_path)
    except Exception as e:
        raise ValueError(f"Failed to decode and extract the base64 archive: {e}")

for i, entry in enumerate(dataset):
    repo_path = os.path.join("websites", f"website_{i}")
    os.makedirs(repo_path, exist_ok=True)
    extract_repo(entry['structure'], repo_path)
    print(f"Extracted repo {i} to {repo_path}")
```

```python
urls = []
for i in range(900):
    port: int = 50000 + i
    urls.append("http://localhost:" + str(port) + "/")

crawled_urls = []
import json
crawled = json.load(open("../Crawler/storage/key_value_stores/default/dataset.json"))

for entry in crawled:
    crawled_urls.append(entry["url"])

# get non-crawled urls
non_crawled_urls = []
for url in urls:
    if url not in crawled_urls:
        non_crawled_urls.append(url)

print(non_crawled_urls)

with open('missed-urls.txt', 'w') as file:
    indices = []
    for url in non_crawled_urls:
        # extract port
        port = int(url.split(":")[-1].split("/")[0])
        indices.append(port - 4000)
        file.write("http://localhost:" + str(port + 46000) + "/\n")

print(indices)
```
